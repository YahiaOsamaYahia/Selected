```python
from sklearn.model_selection import train_test_split
#train 70, test 15, validation 15
x=df.iloc[:,0:-1] #iloc: p ignore coloumn name (library from pandas)
y=df ['deposit']
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.15, shuffle=True, random_state=0)
x_valid, x_test, y_valid, y_test = train_test_split(x_test, y_test, test_size=0.15,shuffle=True, random_state=0)
```

```python
#standization (Many machine learning algorithms perform better when numerical input variables are scaled to a standard range.) 
#Standardization scales each input variable separately by subtracting the mean (called centering) and dividing by the standard deviation to shift the distribution to have a mean of zero and a standard deviation of one[0:1]
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.fit_transform(x_test)
x_valid=sc.fit_transform(x_valid)
```

```python
#SVM
! pip install tensorflow
from tensorflow import keras
from keras.models import Sequential
from sklearn.svm import SVC #Support Vector Machines, one of the most popular and widely used supervised
from sklearn.metrics import accuracy_score, classification_report
from sklearn.model_selection import cross_val_score

#fitting
svc=SVC(kernel="sigmoid", gamma=0.1) #gamma is a parameter, which ranges from 0 to 1. A higher value of gamma will perfectly fit the training dataset, which causes over-fitting. Gamma=0.1 is considered to be a good default value.
svc.fit(x_train ,y_train)

#evaluation 

y_pred_svc = svc.predict(x_test)

class_rep_svc = classification_report(y_test,y_pred_svc)
print("SVC:", class_rep_svc )

svc.score(x_test,y_test)
print("Train Score",svc.score(x_train,y_train))

test_score=accuracy_score(y_test, y_pred_svc)
print("Test Accuracy: " , test_score )
```

```python
#classification report to train
from sklearn.metrics import classification_report, plot_confusion_matrix,plot_roc_curve
print (classification_report(y_train,svc.predict(x_train)))
```

```python
#classification report to test
print (classification_report(y_test,svc.predict(x_test)))
```

```python
#rock curve
fig=plot_roc_curve(svc,x_test,y_test)
plt.title("ROC CURVE")
plt.show()
```

```python
#Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,svc.predict(x_test))
cm
```

```python
#learning curve
from sklearn.datasets import load_digits
from sklearn.model_selection import learning_curve
x=x_train
y=y_train
train_sizes, train_scores, test_scores, fit_times, _ = learning_curve(svc, x, y, cv=30,return_times=True)
plt.plot(train_sizes,np.mean(train_scores,axis=1))
```
