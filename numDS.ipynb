```python
X = df.loc[:, df.columns!='loan']
Y = df['loan']
```

```python
X
```

```python
Y.shape
```

```python
from sklearn.model_selection import train_test_split
#Spilt data to 3 Part(Train,Validation,Test)
 
X_train, X_test, y_train, y_test = train_test_split(
    X, 
    Y, 
    test_size=0.15, 
    shuffle=True,
    random_state=0)
```

```python
X_train, X_valid, y_train, y_valid = train_test_split(
    X_train, 
    y_train, 
    test_size=0.15, 
    shuffle=True,
    random_state=42
)
```

```python
# #standization 
import sklearn
from sklearn.preprocessing import StandardScaler
from keras.callbacks import EarlyStopping
import keras
from keras.models import Sequential
from keras.layers import Dense , Flatten

```

```python
ann = tf.keras.models.Sequential()
```

```python
ann.add(tf.keras.layers.Dense(units=100, activation='relu', input_dim=25))
ann.add(Flatten())
```

```python
ann.add(tf.keras.layers.Dense(units=100, activation='relu'))
ann.add(Flatten())
```

```python
ann.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))
ann.add(Flatten())
```

```python
ann.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
history = ann.fit(X_train, y_train, epochs = 100, batch_size = 10, verbose = 0)
```

```python
train_loss ,acc= ann.evaluate(X_train, y_train)
print('loss score:', train_loss)
print('Train accuracy:', acc)
```

```python
Y_pred=ann.predict(X_valid)
```

```python
import tensorflow as tf

m = tf.keras.metrics.BinaryAccuracy()
m.update_state(y_valid,Y_pred)
print('Final Accuracy score: ', m.result().numpy()) 
```

```python
loss_values = history.history['loss']
epochs = range(1, len(loss_values)+1)

plt.plot(epochs, loss_values, label='Training Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

plt.show()
```

```python
Y_pred = np.argmax(Y_pred, axis=1)
```

```python
from sklearn.metrics import classification_report, plot_confusion_matrix,plot_roc_curve
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_valid,Y_pred)
cm
```
